```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data = pd.read_csv('creditcard.csv')
credit_data.head()
```

```python
credit_data.info()
```

```python
credit_data.isnull().sum()
```

```python
credit_data['Class'].value_counts()
```

```python
legit = credit_data[credit_data['Class'] == 0]
fraud = credit_data[credit_data['Class'] == 1]
```

```python
print(legit.shape,fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_df= pd.concat([legit_sample,fraud],axis= 0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
x=new_df.drop(columns='Class',axis =1)
y = new_df['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size= 0.2,random_state=2,stratify= y)
```

```python
print(X_test.shape , X_train .shape , Y_test.shape , Y_train.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_Predition = model.predict(X_train)
Trainig_Accuracy = accuracy_score(X_train_Predition,Y_train)
print('Accuracy Score on Training :',Trainig_Accuracy)
```

```python
X_test_Predition = model.predict(X_test)
Testing_Accuracy = accuracy_score(X_test_Predition,Y_test)
print('Accuracy Score on testing :',Testing_Accuracy)
```
